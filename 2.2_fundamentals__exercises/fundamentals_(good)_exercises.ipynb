```python
fruits = [("Apple", 0.99), ("Banana", 0.59), ("Orange", 1.29)]




#{name:<10.10s} writes the name 10 spaces aligned to the left
#{price:>5.2f} writes the price 5 spaces aligned to the right

for name, price in fruits:
    print(f"{name:<10.10s} ${price:>5.2f}")
```

### Exercise 2

```python
score = input("Enter the student's score:")

#isdigit() ensures that grade(x) only takes integers as input

if score.isdigit() == False:
    print("Error, invalid score")
else:
    grade(int(score))





def grade(x):
    grade_letter = ''
    if x > 100 or x < 0:
        return "Error, invalid score"
    elif x >= 70:
        grade_letter = "A"
    elif x >= 50:
        grade_letter = "B"
    elif x >= 35:
        grade_letter = "C"
    elif x >= 0:
        return "Fail"
    print(f'The student\'s letter grade is: {grade_letter}')







```

### Exercise 3

```python
def factorial(n):
    if n < 0:
        return "Invalid input"
    elif n == 0:
        return 1
    else:
        result = 1
        for number in range(1, n+1):
            result *= number
        return result


print(factorial(5))   # Testing, expected output: 120
print(factorial(-1))  # Testing, expected output: Invalid input
```
