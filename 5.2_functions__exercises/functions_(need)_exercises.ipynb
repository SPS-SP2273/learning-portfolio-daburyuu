#### Exercise 1

```python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'

greeting("Batman")
```

Line 4 is not indented under the "if" block, and "return" ends the whole function. Hence, if the name is "Batman", it returns the statement in the "if" loop and ends the function, but if the name is not "Batman", then the "if" block is skipped and line 4 is run instead. As a result, the code works as expected despite not having an "else" statement.

#### Exercise 2

```python
def calculate_bmi(weight, height):
    BMI = weight / (height ** 2)
    if BMI < 18.5:
        return "Underweight"
    elif BMI >= 18.5 and BMI <= 24.9:
        return "Normal weight"
    elif BMI >= 25 and BMI <= 29.9:
        return "Overweight"
    elif BMI >= 30:
        return "Obese"

```

#### Exercise 3

```python
def factorial(n):
    value = 1
    if n == 0:
        return 1
    elif n < 0:
        return "Error, negative integers not allowed"
    for i in range (1, n + 1):
        value *= i
    return value

```
