### Exercise 3

```python
squares = [i**2 for i in range(10)]
print(squares)
```

```python
squares = list()
for i in range(10):
    squares.append(i**2)

print(squares)
```

```python
%%timeit
squares = [i**2 for i in range(10)]
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

The first code took 16 more nanoseconds on average to run, and the first code also has a smaller standard deviation 128.49 nanoseconds, meaning the spread of the times taken was smaller for the first code. To get this result, the code was run 1 million times before stopping to calculate the mean and standard deviation, then this was repeated 6 more times (7 in total) before finding the final mean and standard deviations.
