#### Exercise 1

```python
import numpy as np

temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]

for x in np.array(temperatures_celsius):
    fahrenheit = x * 1.8 + 32
    print(x, fahrenheit)

```

#### Exercise 2

```python
five = [1, 2, 3, 4, 5]

x = 0

# Add the count column, then the colon 2 spaces to the right since tab is 4 spaces
for row_number in five:
    print(f"{row_number}   :", end="\t")

    # Multiply row number by column number, then add a tab for spacing
    for column_number in (five):
        print(f"{row_number * column_number}", end="\t")
    print("\n")
```

```python

# Checks if random number is more than 0.5, and counts it as heads if it is
for expt_number in range(1, 11):
    x = np.random.rand(10)
    heads = x > 0.5
    print(f"Experiment {expt_number}: No. of Heads =", int(np.sum(x[heads])))


```
