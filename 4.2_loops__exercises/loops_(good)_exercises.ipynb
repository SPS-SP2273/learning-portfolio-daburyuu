#### Exercise 1

```python
odd_squares = []

for number in range(0, 10):
    # Skips printing if number is even
    if number % 2 == 0:
        continue
    odd_squares.append(number ** 2)
```

#### Exercise 2

```python
# If number is odd, return the square

[number ** 2 for number in range(10) if number % 2 == 1]
```

#### Exercise 3

```python
%%timeit

odd_squares = []

for number in range(0, 10):
    # Skips printing if number is even
    if number % 2 == 0:
        continue
    odd_squares.append(number ** 2)

```

```python
%%timeit

# If number is odd, return the square

[number ** 2 for number in range(10) if number % 2 == 1]

```

The second one is roughly 71 ns faster than the first one.

#### Exercise 5

```python
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]


# Start range with 1 because (n + 1) out of range, so use (n - 1) instead

for n in range (1, len(co2_data)):
    (prev_year, prev_co2) = co2_data[n - 1]
    (this_year, this_co2) = co2_data[n]
    if prev_co2 <= this_co2 - 3:
        co2_diff = this_co2 - prev_co2
        # Comparing 2 floats, so need to round the value
        print(f"{this_year}, {round(co2_diff, 2)}")

```
