#### Exercise 5

```python
x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]

x_1_array = np.array(x_1)
x_2_array = np.array(x_2)


print(x_1_array.mean(), "x_1 mean")
print(x_2_array.mean(), "x_2 mean")


print(x_1_array.std(), "x_1 std")
print(x_2_array.std(), "x_2 std")


# Split the equation up to make it easier to debug
numerator = np.sum((x_1_array - x_1_array.mean())*(x_2_array - x_2_array.mean()))
denominator_x = np.sum((x_1_array - x_1_array.mean())**2)
denominator_y = np.sum((x_2_array - x_2_array.mean())**2)
denominator = np.sqrt(denominator_x)*np.sqrt(denominator_y)

print(numerator / denominator, "Pearson correlation coefficient r")


np.corrcoef(x_1_array, x_2_array)
```
