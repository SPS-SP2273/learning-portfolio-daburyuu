Task 2

```python
months = {
         'jan': '01', 'feb': '02', 'mar': '03',
         'apr': '04', 'may': '05', 'jun': '06',
         'jul': '07', 'aug': '08', 'sep': '09',
         'oct': '10', 'nov': '11', 'dec': '12'
         }


def rename_my_file(old_file_name):
    only_file_name = os.path.basename(old_file_name)
    split_parts = only_file_name.split("-")
    text_month = split_parts[0]
    text_date_city = only_file_name.split("-")[1]
    
    new_text_name = f"{months[text_month]}-{text_date_city}"

    return new_text_name


print(rename_my_file("oct-08_singapore.txt")) 

```

Task 5

```python
list_cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']


try:
    os.mkdir("Cities")
except FileExistsError:
    print("Cities file created")

# Running code twice runs into error, hence try-except used
for city_name in list_cities:
    path_for_cities = os.path.join("Cities", city_name)
    try:
        os.mkdir(path_for_cities)
        print(f"{path_for_cities} created")
    except FileExistsError:
        print(f"{path_for_cities} already exists")


```

```python
list_cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']


for city_name in list_cities:
    path_to_city_folder = os.path.join("Cities", city_name)

    for x in glob.glob(f"*/*{city_name}.*"):
        print(x)
        try:
            shutil.move(x, path_to_city_folder)
        except:
            os.remove(x)


```
